```python
import matplotlib.pyplot as plt

# Datos
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

categories = ['A', 'B', 'C']
values = [7, 3, 5]

# Grfico de Lnea y Dispersin
plt.figure(figsize=(10, 5))

# Grfico de Lnea
plt.subplot(1, 2, 1)
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Grfico de Lnea')
plt.xlabel('X')
plt.ylabel('Y')

# Grfico de Dispersin
plt.subplot(1, 2, 2)
plt.scatter(x, y, color='r', marker='o')
plt.title('Grfico de Dispersin')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()

# Grfico de Barras y Barras Horizontales
plt.figure(figsize=(10, 5))

# Grfico de Barras
plt.subplot(1, 2, 1)
plt.bar(x, y, color='g', alpha=0.7)
plt.title('Grfico de Barras')
plt.xlabel('X')
plt.ylabel('Y')

# Grfico de Barras Horizontales
plt.subplot(1, 2, 2)
plt.barh(x, y, color='orange', alpha=0.7)
plt.title('Grfico de Barras Horizontales')
plt.xlabel('Y')
plt.ylabel('X')

plt.tight_layout()
plt.show()

# Grfico de rea
plt.figure(figsize=(8, 5))
plt.fill_between(x, y, color='purple', alpha=0.5)
plt.title('Grfico de rea')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'lightblue'])
plt.title('Grfico de Pastel')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.stackplot(x, y, labels=['A', 'B', 'C'], colors=['lightcoral', 'lightgreen', 'lightblue'])
plt.title('Grfico de rea Apilada')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
import seaborn as sns

# Datos
categories = ['A', 'B', 'C']
values = [7, 3, 5]

plt.figure(figsize=(8, 5))
sns.violinplot(x=categories, y=values, palette='pastel')
plt.title('Grfico de Violn')
plt.xlabel('Categoras')
plt.ylabel('Valores')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
categories = ['A', 'B', 'C']
values = [7, 3, 5]

# Datos adicionales para el grfico de radar
num_categories = len(categories)
angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False)
values += values[:1]  # Cerrar el crculo

# Ajustar el nmero de ngulos para coincidir con el nmero de categoras
angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False)
values = np.concatenate((values, [values[0]]))

plt.figure(figsize=(8, 8))
plt.polar(angles, values, 'o-', color='purple', linewidth=2)
plt.fill(angles, values, alpha=0.25, color='purple')
plt.thetagrids(angles * 180/np.pi, categories)
plt.title('Grfico de Radar')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
bubble_sizes = [30, 80, 150, 200, 250]  # Tamaos de las burbujas

plt.figure(figsize=(10, 5))
plt.scatter(x, y, s=bubble_sizes, alpha=0.7)
plt.title('Grfico de Burbujas')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import seaborn as sns

# Datos
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.figure(figsize=(8, 5))
sns.kdeplot(x, y, cmap='viridis', fill=True)
plt.title('Grfico de Densidad')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
