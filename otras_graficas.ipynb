```python
import matplotlib.pyplot as plt

# Datos
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

categories = ['A', 'B', 'C']
values = [7, 3, 5]

# Gráfico de Línea y Dispersión
plt.figure(figsize=(10, 5))

# Gráfico de Línea
plt.subplot(1, 2, 1)
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Gráfico de Línea')
plt.xlabel('X')
plt.ylabel('Y')

# Gráfico de Dispersión
plt.subplot(1, 2, 2)
plt.scatter(x, y, color='r', marker='o')
plt.title('Gráfico de Dispersión')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()

# Gráfico de Barras y Barras Horizontales
plt.figure(figsize=(10, 5))

# Gráfico de Barras
plt.subplot(1, 2, 1)
plt.bar(x, y, color='g', alpha=0.7)
plt.title('Gráfico de Barras')
plt.xlabel('X')
plt.ylabel('Y')

# Gráfico de Barras Horizontales
plt.subplot(1, 2, 2)
plt.barh(x, y, color='orange', alpha=0.7)
plt.title('Gráfico de Barras Horizontales')
plt.xlabel('Y')
plt.ylabel('X')

plt.tight_layout()
plt.show()

# Gráfico de Área
plt.figure(figsize=(8, 5))
plt.fill_between(x, y, color='purple', alpha=0.5)
plt.title('Gráfico de Área')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'lightblue'])
plt.title('Gráfico de Pastel')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.stackplot(x, y, labels=['A', 'B', 'C'], colors=['lightcoral', 'lightgreen', 'lightblue'])
plt.title('Gráfico de Área Apilada')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
import seaborn as sns

# Datos
categories = ['A', 'B', 'C']
values = [7, 3, 5]

plt.figure(figsize=(8, 5))
sns.violinplot(x=categories, y=values, palette='pastel')
plt.title('Gráfico de Violín')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
bubble_sizes = [30, 80, 150, 200, 250]  # Tamaños de las burbujas

plt.figure(figsize=(10, 5))
plt.scatter(x, y, s=bubble_sizes, alpha=0.7)
plt.title('Gráfico de Burbujas')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import seaborn as sns

# Datos
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.figure(figsize=(8, 5))
sns.kdeplot(x, y, cmap='viridis', fill=True)
plt.title('Gráfico de Densidad')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import seaborn as sns

# Datos
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

plt.figure(figsize=(8, 5))
sns.heatmap(matrix, annot=True, cmap='coolwarm')
plt.title('Gráfico de Mapa de Calor')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Datos
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Crear una figura tridimensional
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Crear la superficie tridimensional con degradado de colores
surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k', linewidth=0.5)

# Añadir barra de colores
fig.colorbar(surface, shrink=0.5, aspect=10)

# Configuración adicional
ax.set_title('Gráfico 3D con Degradado de Colores')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(elev=20, azim=-45)  # Ajustar la vista

plt.show()
```
